```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
test_log = pd.read_csv('first_working.csv', delimiter=";")
test_log.head()
```

```python
sns.scatterplot(
    data=test_log,
    x="timestamp",
    y="profit_and_loss",
    hue="product"  # this separates by product
)

average_profit = test_log["profit_and_loss"].mean()
plt.axhline(y=average_profit, color='red', linestyle='--', alpha=0.3, label="Overall Avg")

plt.title("Profit and Loss by Product Over Time")
plt.xlabel("Timestamp")
plt.ylabel("Profit and Loss")
plt.legend()
plt.show()
```

```python
# Filter for RAINFOREST_RESIN
resin_log = test_log[test_log["product"] == "RAINFOREST_RESIN"]

# Plot
sns.scatterplot(
    data=resin_log,
    x="timestamp",
    y="profit_and_loss"
)

# Add average profit line
average_profit_resin = resin_log["profit_and_loss"].mean()
plt.axhline(y=average_profit_resin, color='blue', linestyle='--', alpha=0.3, label="RAINFOREST_RESIN Avg")

# Labels and title
plt.title("Profit and Loss for RAINFOREST_RESIN Over Time")
plt.xlabel("Timestamp")
plt.ylabel("Profit and Loss")
plt.legend()
plt.show()
```

```python
# Filter for RAINFOREST_RESIN
resin_log = test_log[test_log["product"] == "KELP"]

# Plot
sns.scatterplot(
    data=resin_log,
    x="timestamp",
    y="profit_and_loss"
)

# Add average profit line
average_profit_resin = resin_log["profit_and_loss"].mean()
plt.axhline(y=average_profit_resin, color='blue', linestyle='--', alpha=0.3, label="KELP Avg")

# Labels and title
plt.title("Profit and Loss for KELP Over Time")
plt.xlabel("Timestamp")
plt.ylabel("Profit and Loss")
plt.legend()
plt.show()
```
