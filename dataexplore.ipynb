```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
test_log = pd.read_csv('test_log.csv', delimiter=";")
test_log.head()
```

```python
sns.scatterplot(y =test_log["profit_and_loss"], x= test_log["timestamp"])
average_profit = sum(test_log["profit_and_loss"]/len(test_log['profit_and_loss']))

plt.axhline(y=average_profit, color = 'red', alpha = 0.2)
```
